```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import casadi as cs
import torch
sys.path.append('..')
```

```python
# pendulum from as gym
m = 1
l = 1
k = 0
g = 10
nx = 2
nu = 1
u = cs.SX.sym('u', nu)
x = cs.SX.sym('x', nx)
dx1 = x[1]
dx2 = -g/l * cs.sin(x[0]) - k/m * x[1] + 1/(m*l**2) * u
dx = cs.vertcat(dx1, dx2)
dt = 0.05
cc_order = 4

system = cs.Function("sys", [x, u], [dx])
ode = {'x': x, 'ode': dx, 'p': u}
opts = {'tf': dt}
ode_solver = cs.integrator('F', 'idas', ode, opts)
```

```python
# odeSolver = SpringSlider_ode_solver()
basepath = os.path.dirname(os.getcwd())
```

```python
modelpath = basepath + '/Models/TrainedOfflineModel/1'
ABC = torch.load(modelpath + '/ABC.pth').cpu().eval()
EnNet = torch.load(modelpath + '/EnNet.pth').cpu().eval()
ABCweight = list(ABC.named_parameters())
```

## Eigen value

```python
A = ABCweight[0][1].detach().numpy()
E,V = np.linalg.eig(A)
import matplotlib.pyplot as plt
plt.figure(figsize=[7,7])
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
for e in E:
    ax.scatter(e.real, e.imag, c='r')
circle = plt.Circle((0, 0), 1.0, color='b', fill=False)
plt.gca().add_patch(circle)
plt.savefig("/home/jiayun/Desktop/systemEigenvalue.jpg", dpi=200)
```

## Vector field

```python
x1 = np.linspace(-2*np.pi,2*np.pi,30)
x2 = np.linspace(-2*np.pi,2*np.pi,30)
X0 = []
X = []
LearnX = []

for i, x1i in enumerate(x1):
    for j, x2i in enumerate(x2):
        x0_1, x0_2 = np.cos(x1i), np.sin(x1i)
        Learnx0 = np.array([x0_1, x0_2, x2i]).reshape(3,1)
        x0 = np.array([x1i, x2i]).reshape(2,1)
        res = ode_solver(x0=x0, p=0)
        xnext = res['xf'].full()
        X0.append(x0)
        X.append(xnext)
        
        ##############
        Learnx0 = Learnx0.squeeze()
        x0_tensor = torch.from_numpy(Learnx0).cpu().float()
        x0_lifting_tensor = EnNet(x0_tensor)
        xnext_lifting_tensor = ABC.A @ x0_lifting_tensor
        xnext_tensor = ABC.C @ xnext_lifting_tensor
        xnext_learn = xnext_tensor.detach().numpy()
        xnext_learn = xnext_learn[:, np.newaxis]
        LearnX.append(xnext_learn)
        
        
X = np.concatenate(X, axis=1).T
X0 = np.concatenate(X0, axis=1).T
LearnX = np.concatenate(LearnX, axis=1).T
```

```python
plt.figure(figsize=[10,8])

for i, target in enumerate(X):
    start = X0[i]
    Ltarget = LearnX[i]
    x, y = start[0], start[1]
    dx = (x - np.cos(target[0])) * 0.15
    dy = (y - target[1]) * 0.15
    dxL = (x - Ltarget[0]) * 0.15
    dyL = (y - Ltarget[2]) * 0.15
    plt.arrow(x,y,dx,dy,head_width=.05, color='k')
    plt.arrow(x,y,dxL,dyL,head_width=.05, color='r')
plt.arrow(x,y,dx,dy,head_width=.05, color='k', label='Real ODE')
plt.arrow(x,y,dxL,dyL,head_width=.05, color='r', label='Koopman')
plt.xlabel(r"$cos(\theta)$")
plt.ylabel(r"$\dot{\theta}$")
plt.legend()
plt.savefig("/home/jiayun/Desktop/systemVectorField.jpg", dpi=200)
plt.show()
```
