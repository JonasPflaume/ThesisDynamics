```python
import sys
sys.path.append('..')
from Common.Dataset_generator import Datasets
import numpy as np
from regression import CV_RR, BoostRegressor
from regression_utils import build_dataset_RR
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
D = Datasets('../data/trajectories/', 1, False, False)
D_t = Datasets('../data/test_trajectory/', 1, False, False)
```

```python
Y, X = build_dataset_RR(D, 1)
Y_t, X_t = build_dataset_RR(D_t, 1)
```

```python
Y.shape
```

```python
def feature(X):
    aug = X
    poly = PolynomialFeatures(3)
    X = poly.fit_transform(aug)
    return X
```

```python
labd = np.logspace(1,7,10)
RR = CV_RR(labd)
```

```python
RR.fit(feature(X), Y, 4)
```

```python
BR = BoostRegressor(1)
BR.fit(feature(X), Y[:, 5])
```

### Active GPR

```python
def GP_regression_std(regressor, X):
    _, std = regressor.predict(X, return_std=True)
    query_idx = np.argmax(std)
    return query_idx, X[query_idx], np.max(std)
from modAL.models import ActiveLearner
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
```

```python
n_initial = 10
initial_idx = np.random.choice(range(len(X)), size=n_initial, replace=False)
X_training, Y_training = X[initial_idx], Y[initial_idx]
kernel = RBF(length_scale = 1)
regressor = ActiveLearner(
    estimator=GaussianProcessRegressor(kernel=kernel),
    query_strategy=GP_regression_std,
    X_training=X_training.reshape(-1, 21), y_training=Y_training.reshape(-1, 7)
)
```

```python
error_l = []
import tqdm
for _ in tqdm.tqdm(range(2000)):
    query_idx, query_instance, max_std = regressor.query(X)

    regressor.teach(X[query_idx].reshape(1, 21), Y[query_idx].reshape(1, 7))
    pred = regressor.predict(X_t)
    error = np.linalg.norm(pred[:, 5] - Y_t[:, 5])
    error_l.append(error)
    print("error: %.3f" % error)
```

```python
regressor.X_training.shape
```

### Evaluate

```python
pred = RR.predict(feature(X_t), RR.beta)
print(np.linalg.norm(pred[:, 5] - Y_t[:, 5]))
```
