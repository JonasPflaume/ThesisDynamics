```python
def InertiaLoss(xx, yy, zz, xy, xz, yz): #L_1xx, L_1yy, L_1zz, L_1xy, L_1xz, L_1yz
    I = [[xx,xy,xz],[xy,yy,yz],[xz,yz,zz]]
    I = np.array(I)
    e,v = np.linalg.eig(I)
    L = np.trace(I)/2 - max(e)
    return L
```

```python
def fitness(x, mu = 100):
    
    total_error = 0
    
    m_1, m_2, m_3, m_4, m_5, m_6, m_7 = x[0:7]
    l_1x, l_1y, l_1z = x[7:10]
    l_2x, l_2y, l_2z = x[10:13]
    l_3x, l_3y, l_3z = x[13:16]
    l_4x, l_4y, l_4z = x[16:19]
    l_5x, l_5y, l_5z = x[19:22]
    l_6x, l_6y, l_6z = x[22:25]
    l_7x, l_7y, l_7z = x[25:28]
    L_1xx, L_1yy, L_1zz, L_1xy, L_1xz, L_1yz = x[28:34]
    L_2xx, L_2yy, L_2zz, L_2xy, L_2xz, L_2yz = x[34:40]
    L_3xx, L_3yy, L_3zz, L_3xy, L_3xz, L_3yz = x[40:46]
    L_4xx, L_4yy, L_4zz, L_4xy, L_4xz, L_4yz = x[46:52]
    L_5xx, L_5yy, L_5zz, L_5xy, L_5xz, L_5yz = x[52:58]
    L_6xx, L_6yy, L_6zz, L_6xy, L_6xz, L_6yz = x[58:64]
    L_7xx, L_7yy, L_7zz, L_7xy, L_7xz, L_7yz = x[64:70]

    e1 = L_1zz + L_2yy - 0.0289907
    e2 = L_2xx - L_2yy + L_3yy + 79*l_3z/125 + 6241*m_3/62500 + 426649*m_4/4000000 + \
                426649*m_5/4000000 + 426649*m_6/4000000 + 426649*m_7/4000000 - 1.0651475
    e3 = L_2xy- -0.0053347
    e4 = L_2xz - 0.0284135
    e5 = L_2yz - -0.0034973
    e6 = L_2zz + L_3yy + 79*l_3z/125 + 6241*m_3/62500 + 426649*m_4/4000000 + 426649*m_5/4000000 +\
                    426649*m_6/4000000 + 426649*m_7/4000000 - 1.1260740
    e7 = l_2x - -0.0057238
    e8 = l_2y - l_3z - 79*m_3/250 - 79*m_4/250 - 79*m_5/250 - 79*m_6/250 - 79*m_7/250 - -3.3476041
    e9 = L_3xx - L_3yy + L_4yy - 1089*m_4/160000 - 0.0105830
    e10 = L_3xy + 33*l_4z/400 - -0.0002799
    e11 = L_3xz - -0.0106875
    e12 = L_3yz - -0.0047958
    e13 = L_3zz + L_4yy + 1089*m_4/160000 + 1089*m_5/80000 + 1089*m_6/80000 + 1089*m_7/80000 - 0.1274740
    e14 = l_3x + 33*m_4/400 + 33*m_5/400 + 33*m_6/400 + 33*m_7/400 - 0.7511272
    e15 = l_3y - l_4z - 0.0237469
    e16 = L_4xx - L_4yy + L_5yy + 96*l_5z/125 + 562599*m_5/4000000 + 562599*m_6/4000000 + 562599*m_7/4000000 - 0.6421652
    e17 = L_4xy + 33*l_5z/400 + 99*m_5/3125 + 99*m_6/3125 + 99*m_7/3125 - 0.1749839
    e18 = L_4xz - 0.0047464
    e19 = L_4yz - -0.0026809
    e20 = L_4zz + L_5yy + 96*l_5z/125 + 617049*m_5/4000000 + 617049*m_6/4000000 + 617049*m_7/4000000 - 0.7572770
    e21 = l_4x - 33*m_5/400 - 33*m_6/400 - 33*m_7/400 - -0.5533095
    e22 = l_4y + l_5z + 48*m_5/125 + 48*m_6/125 + 48*m_7/125 - 2.0172931
    e23 = L_5xx - L_5yy + L_6yy + 121*m_7/15625 - 0.0382051
    e24 = L_5xy - -0.0037326
    e25 = L_5xz - -0.0061429
    e26 = L_5yz - 0.0077041
    e27 = L_5zz + L_6yy + 121*m_7/15625 - 0.0241165
    e28 = l_5x - -0.0100414
    e29 = l_5y - l_6z - 0.0770350
    e30 = L_6xx - L_6yy + L_7yy - 121*m_7/15625 - 0.0024295
    e31 = L_6xy + 11*l_7z/125 - 0.0147750
    e32 = L_6xz - -0.0000140
    e33 = L_6yz - 0.0008204
    e34 = L_6zz + L_7yy + 121*m_7/15625 - 0.0461065
    e35 = l_6x + 11*m_7/125 - 0.2342840
    e36 = l_6y - l_7z - -0.1745756
    e37 = L_7xx - L_7yy - -0.0032513
    e38 = L_7xy - 0.0011841
    e39 = L_7xz - 0.0023950
    e40 = L_7yz - -0.0005794
    e41 = L_7zz - 0.0029619
    e42 = l_7x - -0.0015551
    e43 = l_7y - -0.0004662
    e = [e1, e2, e3, e4, e5, e6, e7, e8, e9, e10, e11, e12, e13, e14, e15, e16, e17, e18, e19,\
         e20, e21, e22, e23, e24, e25, e26, e27, e28, e29, e30, e31, e32, e33, e34, e35, e36,\
         e37, e38, e39, e40, e41, e42, e43]
    e = np.array(e).reshape(43,1)
    # retrieval error
    error1 = e.T @ e
    total_error += error1[0]
    # mass constraints
    M = m_1+m_2+m_3+m_4+m_5+m_6+m_7
    error2 = 0
    if M < 17.5:
        error2 += mu * (17.5 - M) ** 2
    if M > 20:
        error2 += mu * (M - 20) ** 2
    total_error += error2
    # inertia constraints
    error3 = 0
    I_L1 = InertiaLoss(L_1xx, L_1yy, L_1zz, L_1xy, L_1xz, L_1yz)
    I_L2 = InertiaLoss(L_2xx, L_2yy, L_2zz, L_2xy, L_2xz, L_2yz)
    I_L3 = InertiaLoss(L_3xx, L_3yy, L_3zz, L_3xy, L_3xz, L_3yz)
    I_L4 = InertiaLoss(L_4xx, L_4yy, L_4zz, L_4xy, L_4xz, L_4yz)
    I_L5 = InertiaLoss(L_5xx, L_5yy, L_5zz, L_5xy, L_5xz, L_5yz)
    I_L6 = InertiaLoss(L_6xx, L_6yy, L_6zz, L_6xy, L_6xz, L_6yz)
    I_L7 = InertiaLoss(L_7xx, L_7yy, L_7zz, L_7xy, L_7xz, L_7yz)
    if I_L1 < 0:
        error3 -= I_L1
    if I_L2 < 0:
        error3 -= I_L2
    if I_L3 < 0:
        error3 -= I_L3
    if I_L4 < 0:
        error3 -= I_L4
    if I_L5 < 0:
        error3 -= I_L5
    if I_L6 < 0:
        error3 -= I_L6
    if I_L7 < 0:
        error3 -= I_L7
    total_error += mu*error3
    return total_error
```

```python
res_l = []
for _ in range(10):
    x0 = [np.random.uniform(low=LB[i], high=UB[i]) for i in range(70)]
    sigma = 0.03    # initial standard deviation to sample new solutions
    x, es = cma.fmin2(fitness, x0, sigma, {'bounds': [LB, UB]})
    res_l.append(x[:,np.newaxis])
```
