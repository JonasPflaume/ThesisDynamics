```python
import casadi as cs
import numpy as np
```

```python
# Instrantiation (x, dx, u, dt, cc_degree, Q, R, N, ub_x, lb_x, ub_u, lb_u)
# main loop (N_sim, ref_traj, x_0)
# pendulum example
m = 1
l = 1
k = 0
g = 10
nx = 2
nu = 1
u = cs.SX.sym('u', nu)
x = cs.SX.sym('x', nx)
dx1 = x[1]
dx2 = -g/l * cs.sin(x[0]) - k/m * x[1] + 1/(m*l**2) * u
dx = cs.vertcat(dx1, dx2)
dt = 0.05
cc_order = 4
```

```python
N = 20
lb_x = -500*np.ones((nx,1))
ub_x = 500*np.ones((nx,1))
# input constraints, at least around 12
lb_u = -2*np.ones((nu,1))
ub_u = 2*np.ones((nu,1))
K = 1/2 * m * (x[1] * l) ** 2
P = (l - cs.cos(x[0])*l) * m * g
stage_cost = - 10*P + K + 0.1 * u
stage_cost_fn = cs.Function('stage_cost', [x, u], [stage_cost])
terminal_cost = - 100*P + K
terminal_cost_fn = cs.Function('ter_cost', [x], [terminal_cost])
```

```python
solver = NLMPC_ODE(x, dx, u, dt, cc_order, stage_cost_fn, terminal_cost_fn, N, ub_x, lb_x, ub_u, lb_u)
```

```python
N_sim = 200
x0 = np.zeros([2,1])
res_U, res_X = solver.MPC_loop(N_sim, x0)
```

```python
import matplotlib.pyplot as plt
plt.plot(res_X.T)
plt.xlabel('horizon')
plt.ylabel('angle')
plt.grid()
plt.show()
```

```python
plt.plot(res_U.T)
plt.xlabel('horizon')
plt.ylabel('torque')
plt.show()
```

```python
import gym
env = gym.make('Pendulum-v0')
```

```python
m = 1
l = 1
k = 0
g = 10
nx = 2
nu = 1
u = cs.SX.sym('u', nu)
x = cs.SX.sym('x', nx)
dx1 = x[1]
dx2 = 3*g/(2*l) * cs.sin(x[0]) + k/m * x[1] + 3/(m*l**2) * u
dx = cs.vertcat(dx1, dx2)
dt = 0.05
cc_order = 4
```

```python
N = 20
lb_x = -500*np.ones((nx,1))
ub_x = 500*np.ones((nx,1))
# input constraints, at least around 12
lb_u = -2*np.ones((nu,1))
ub_u = 2*np.ones((nu,1))
K = 1/2 * m * (x[1] * l) ** 2
P = (l + cs.cos(x[0])*l) * m * g
stage_cost = - 10*P + K + 0.01 * u
stage_cost_fn = cs.Function('stage_cost', [x, u], [stage_cost])
terminal_cost = - 10*P + K
terminal_cost_fn = cs.Function('ter_cost', [x], [terminal_cost])
solver_gym = NLMPC_ODE(x, dx, u, dt, cc_order, stage_cost_fn, terminal_cost_fn, N, ub_x, lb_x, ub_u, lb_u)
```

```python
import tqdm
done = False
obs = env.reset()
theta = np.arccos(obs[0])
thetaD = obs[2]
x = np.array([theta, thetaD]).reshape(-1,1)
res = [x]
for i in tqdm.tqdm(range(200)):
    action = solver_gym.MPC_step(x)
    obs, _, done,_ = env.step(action)
    env.render()
    theta = np.arccos(obs[0])
    thetaD = obs[2]
    x = np.array([theta, thetaD]).reshape(-1,1)
    res.append(x)
```

```python
res_ = np.concatenate(res, axis=1)
plt.plot(res_[1,:], '-r')
plt.plot(res_[0,:])
```
