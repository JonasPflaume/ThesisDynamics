#### Preprocessing

```python
import os

def trans_format(traj_raw):
    # traj_raw: the raw data generated by matlab
    with open('./trajectories/Trajectory_{}/gen_q0.txt'.format(traj_raw), 'r') as F:
        q0 = F.readlines()
        q0 = list(map(str.strip, q0))
        q0 = np.array(q0, dtype=float).reshape(1, 7)
        
    with open('./trajectories/Trajectory_{}/gen_dq.txt'.format(traj_raw), 'r') as F:
        dq = []
        for d_q in F:
            dq.append(d_q.strip())
        dq = np.array(dq, dtype=float)
        dq = dq.reshape(-1, 7)
        
    with open('./trajectories/Trajectory_{}/gen_time.txt'.format(traj_raw), 'r') as F:
        t = []
        for t_ in F:
            t.append(t_.strip())
        t = np.array(t, dtype=float)
        t = t.reshape(-1, 1)
        
    if os.path.isdir('./trajectories/traj{}'.format(traj_raw)):
        pass
    else:
        os.mkdir('./trajectories/traj{}'.format(traj_raw))
    
    with open('./trajectories/traj{}/q0.txt'.format(traj_raw), 'w') as F:
        for item in q0:
            stuff = ' '.join(str(x) for x in item)
            F.write(stuff + '\n')
            
    with open('./trajectories/traj{}/dq.txt'.format(traj_raw), 'w') as F:
        for item in dq:
            stuff = ' '.join(str(x) for x in item)
            F.write(stuff + '\n')
    
    with open('./trajectories/traj{}/t.txt'.format(traj_raw), 'w') as F:
        for item in t:
            stuff = ' '.join(str(x) for x in item)
            F.write(stuff + '\n')
        

for n in range(1,21):
    trans_format(n)
```

```python
def pick_trajectory(traj):
    '''
    function for picking a trajectory
    traj generated from 5 superpositioning harmonics
    
    traj: str
        the folder name of traj, start form traj0 ...
    return
        t - time stamp
        q - joints config
    '''
    
    with open('./trajectories/{}/q0.txt'.format(traj), 'r') as F:
        q0 = re.split(' |\n', F.readline())
        q0 = np.array(q0[:-1], dtype=float)

    with open('./trajectories/{}/t.txt'.format(traj), 'r') as F:
        t = []
        for item in F:
            t.append(item)
        t = np.array(t, dtype=float) + 2
        
    sampling_freq = int(len(t)/(max(t)-min(t))) # hz
    print("sampling frequency is: {} Hz".format(sampling_freq))
    
    with open('./trajectories/{}/dq.txt'.format(traj), 'r') as F:
        dq = []
        for item in F:
            temp = re.split(' |\n', item)
            temp = np.array(temp[:-1], dtype=float)
            dq.append(temp)
        dq = np.array(dq).reshape(len(t), 7)

    q = np.zeros([*dq.shape])
    for idx, dqt in enumerate(dq):
        if idx == 0:
            q[idx,:] = q0
        else:
            q[idx,:] = q[idx-1,:] + dq[idx,:] * 1/sampling_freq

    plt.figure(figsize=[8, 4], dpi=80)
    plt.plot(t, q)
    plt.xlabel('Time [s]')
    plt.ylabel('Joints angular [rad]')
    plt.show()
    return t, q
    
t, q = pick_trajectory('traj20')

### test on simulation:
# ok: 1 2 3 5 6 7 9 10 11 12 13 15 16 17 18 
# skeptical: 19
# hit desk: 4 8 14 20
# succ
```

```python
t = t #* 4
```

```python
bot.setControllerWriteData(2)
bot.move(q, t)

while bot.getTimeToEnd()>0:
    bot.step(C, .1)
    time.sleep(.1)
```

```python
bot.setControllerWriteData(0)
bot.home(C)
```
